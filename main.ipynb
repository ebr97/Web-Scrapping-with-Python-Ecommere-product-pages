```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
baseurl = 'https://www.thewhiskyexchange.com'

# we will overwrite our user-agent
# if there will appear python most lickley will be block
# site: https://developers.whatismybrowser.com/useragents/explore/software_type_specific/web-browser/
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36'}

productlinks = [] # to store all the links
for x in range(1,3): # to loop in all pages
    if x == 1:
        r = requests.get('https://www.thewhiskyexchange.com/c/35/japanese-whisky')
    elif x == 2:
        r = requests.get(f'https://www.thewhiskyexchange.com/c/35/japanese-whisky?pg={x}')
    soup = BeautifulSoup(r.content, 'lxml')
    productlist = soup.find_all('li', class_='product-grid__item')
    for item in productlist:
        link = item.a['href']
        productlinks.append(baseurl+link)
```

```python
################# Retreiving information about the PRODUCT from the product links ##################

#testlink = 'https://www.thewhiskyexchange.com/p/2940/yamazaki-12-year-old'
whiskylist = []
for link in productlinks:
    r = requests.get(link, headers = headers)
    soup = BeautifulSoup(r.content, 'lxml')

    name = soup.find('h1', class_='product-main__name').text.strip()
    price = soup.find('p', class_='product-action__price').text.strip()
    try:
        rating = soup.find('span', class_='review-overview__rating star-rating star-rating--45').text.strip()
    except:
        rating = 'no rating'
    
    #no_reviews = soup.find('span', class_='review-overview__count').text[1:3].strip()
    #reviews = soup.find('span', class_='review-overview__count').text.strip()

    whisky = {
        'name': name,
        'rating': rating,
        #'reviews': no_reviews+' Reviews',
        'price': price
        }
    whiskylist.append(whisky)
    print('Saving: {}'.format(whisky['name']))

df = pd.DataFrame(whiskylist)
df
    
    
```
